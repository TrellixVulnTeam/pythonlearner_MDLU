# If data does not have headers

```python
df5 = pd.read_csv('files/supermarkets_no_headers.csv', header=None)
df5
```

# Set column names

```python
df5.columns = ['ID','Address','City','State','Country','Name','Employees']
df5
```

```python
df5 = pd.read_csv('files/supermarkets_no_headers.csv', header=None, names = ['ID','Address','City','State','Country','Name','Employees'])
df5.set_index('Address', inplace=True, drop=False)
df5
```

# To extract data from data frame
#### Label based indexing use loc

```python
df5.loc['735 Dolores St':'332 Hill St', 'City' : 'Country']
```

```python
df5.iloc[1:3, 2:5]
```

# Deleting Columns or Rows

Note that .drop is not an inplace method

```python
df5.drop('City',1)
```

```python
df5.drop(df5.index[0:3],0) #drops rows zero to 3
```

```python
df5.drop(df5.columns[0:3],1) #drops columns zero to 3
```

# Updating and Adding Columns and Rows

```python
df5['Continent'] = 'North America'
df5
```

```python
df5_t = df5.T #transposes the dataframe
df5_t
```

A trick to add a new row is to transpose the data and then add a new column and then transpose it back. See below

```python
my_address = [7, 'Test Address', 'Toronto', 'Ontario', 'CA','Kyle', 10, 'North America']
df5_t['My Address'] = my_address
df5_t.T
```
