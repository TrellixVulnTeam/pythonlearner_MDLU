# Converting Address to Coordinates Using Geopy

```python
import geopy 
dir(geopy) # Let's look at what's in geopy!
```

## We need geocoders!
Note that to use geocoder you will need an internet connection!

```python
from geopy.geocoders import ArcGIS #There are a few geocoders here but we're going to use ArcGIS (Nominatim has a bug!)
nom = ArcGIS() #we have to store this object into a variable
```

```python
geo=nom.geocode('3995 23rd St, San Francisco, CA 94114')
print(geo.latitude, geo.longitude)
```

```python
df
```

```python
df['Address'] = df['Address'] + ',' + df['City'] + ',' + df['State'] + ',' + df['Country']
df
```

```python
df['Coordinates'] = df['Address'].apply(nom.geocode)
df
```

```python
# We can't simply apply latitude because we have to take the location object to apply latitude, 
# which is why we have to use a lambda function
# So this is saying apply latitude if x is not none, if it is then return none
df['Latitude']=df['Coordinates'].apply(lambda x : x.latitude if x != None else None)


df['Longtitude'] = df['Coordinates'].apply(lambda x : x.longitude if x != None else None)
df
```
